## ***Defining functions of the board***

```python
def get_board(emptyList, randPile, player):
    print("Let's see how the board is looking right now...")
    print("-" * 25)
    emptyList.clear()
    for i in range(randPile):
        randRock = random.randint(1, 6)  #number of rocks will be between 1 to 6
        print('Pile_no {}: {}'.format(i + 1, '/*\\' * randRock))
        emptyList.append(randRock)
    print("-" * 25)
```

## ***Selecting the the number of stones to remove from piles***


```python
def get_valid_input(emptyList, player):
    while True:
        piles = input('From which pile you want to remove stones, {}??: '.format(player)).strip()
        stones = input('How many stones do you want to remove (maximum 3)??: ').strip()

        if (stones and piles) and (stones.isdigit()) and (piles.isdigit()):
            piles = int(piles)
            stones = int(stones)
            if (0 < stones <= 3) and (0 < piles <= len(emptyList)):  #maximum 3 stones can be picked
                if stones <= emptyList[piles - 1]:
                    break
            else:
                print("Invalid input. Please try again, {}.....".format(player))

        print("OOPs!!! You've entered an invalid input. Please try again, {}.....".format(player))

    emptyList[piles - 1] -= stones
```

```python
def minimax(emptyList, depth, isMaximizing):
  #isMaximizing indicates if it is the maxPlayer turn or not
    if emptyList == [0] * len(emptyList):
        return -1 if isMaximizing else 1 #if it is the maxPlayer turn and the list is empty then returns -1, maxPlayer loses the game
        #but if it is minPlayer turn and the list was empty then the maxPlayer will win the game returning 1

    if isMaximizing:
        maxVal = float('-inf')
        for pile in range(len(emptyList)):
            if emptyList[pile] > 0: #checking if the pile is not empty
                for stones in range(1, min(4, emptyList[pile] + 1)): #the loop will start from 1 and as max 3 stones can be removed at once, it will ensure
                #that the loop will not remove more stones than are available in the pile
                    newEmptyList = emptyList.copy() #creating a new list assigning the values of the crnt state
                    newEmptyList[pile] -= stones #removing stones from the pile
                    eval = minimax(newEmptyList, depth + 1, False) #recursively evaluating
                    maxVal = max(maxVal, eval) #updating the max value
        return maxVal
    else:
        minVal = float('inf')
        for pile in range(len(emptyList)):
            if emptyList[pile] > 0:
                for stones in range(1, min(4, emptyList[pile] + 1)):
                    newEmptyList = emptyList.copy()
                    newEmptyList[pile] -= stones
                    eval = minimax(newEmptyList, depth + 1, True)
                    minVal = min(minVal, eval)
        return minVal

def best_move_minimax(emptyList):
    bestMove = None
    bestValue = float('-inf')

    for pile in range(len(emptyList)): #iterating all the piles to find the best move
        if emptyList[pile] > 0:
            for stones in range(1, min(4, emptyList[pile] + 1)):
                newEmptyList = emptyList.copy()
                newEmptyList[pile] -= stones
                moveValue = minimax(newEmptyList, 0, False)
                if moveValue > bestValue:
                    bestValue = moveValue
                    bestMove = (pile + 1, stones)

    return bestMove
```

```python
def alpha_beta(emptyList, depth, alpha, beta, isMaximizing):
    if emptyList == [0] * len(emptyList):
        return -1 if isMaximizing else 1

    if isMaximizing:
        maxVal = float('-inf')
        for pile in range(len(emptyList)):
            if emptyList[pile] > 0:
                for stones in range(1, min(4, emptyList[pile] + 1)):
                    newEmptyList = emptyList.copy()
                    newEmptyList[pile] -= stones
                    eval = alpha_beta(newEmptyList, depth + 1, alpha, beta, False)
                    maxVal = max(maxVal, eval)
                    alpha = max(alpha, eval)
                    if beta <= alpha:
                        break
        return maxVal
    else:
        minVal = float('inf')
        for pile in range(len(emptyList)):
            if emptyList[pile] > 0:
                for stones in range(1, min(4, emptyList[pile] + 1)):
                    newEmptyList = emptyList.copy()
                    newEmptyList[pile] -= stones
                    eval = alpha_beta(newEmptyList, depth + 1, alpha, beta, True)
                    minVal = min(minVal, eval)
                    beta = min(beta, eval)
                    if beta <= alpha:
                        break
        return minVal

def best_move_alpha_beta(emptyList):
    bestMove = None
    bestValue = float('-inf')
    alpha = float('-inf')
    beta = float('inf')

    for pile in range(len(emptyList)):
        if emptyList[pile] > 0:
            for stones in range(1, min(4, emptyList[pile] + 1)):
                newEmptyList = emptyList.copy()
                newEmptyList[pile] -= stones
                moveValue = alpha_beta(newEmptyList, 0, alpha, beta, False)
                if moveValue > bestValue:
                    bestValue = moveValue
                    bestMove = (pile + 1, stones)
                alpha = max(alpha, moveValue)

    return bestMove
```

```python
def ai_move_minimax(emptyList):
    move = best_move_minimax(emptyList)

    if move is not None:
        pile, stones = move
        print("\nNimX removes {} stones from pile {}.\n".format(stones, pile))
        emptyList[pile - 1] -= stones
        print("After NimX's turn,,,")
        continue_game(emptyList, "NimX")
    else:
        print("No valid move found for MINIMAX. This should not happen in a correct implementation.")

def ai_move_alpha_beta(emptyList):
    move = best_move_alpha_beta(emptyList)

    if move is not None:
        pile, stones = move
        print("\nNimX removes {} stones from pile {}.\n".format(stones, pile))
        emptyList[pile - 1] -= stones
        print("After NimX's turn,,,")
        continue_game(emptyList, "NimX")
    else:
        print("No valid move found for Alpha-Beta. This should not happen in a correct implementation.")
```

```python
def ida_star(emptyList):
    threshold = heuristic(emptyList) #initializing the threshold with heuristic function
    path = []

    while True:
        temp = search(emptyList, 0, threshold, path)
        if temp == 'FOUND':
            return path[0]  #return the first move if the search finds the solution
        if temp == float('inf'):
            return None
        threshold = temp #updating threshold with min value greater than crnt threshold

def search(emptyList, gn, threshold, path):
    fn = gn + heuristic(emptyList)
    if fn > threshold:
        return fn
    if emptyList == [0] * len(emptyList):
        return 'FOUND'

    minVal = float('inf') #if the goal is not found in crnt threshold, minVal provides the next threshold

    for pile in range(len(emptyList)):
        if emptyList[pile] > 0:
            for stones in range(1, min(4, emptyList[pile] + 1)):
                newEmptyList = emptyList.copy()
                newEmptyList[pile] -= stones
                path.append((pile + 1, stones))
                temp = search(newEmptyList, gn + 1, threshold, path)
                if temp == 'FOUND':
                    return 'FOUND'
                if temp < minVal:
                    minVal = temp
                path.pop() #removing the last move to backtrack

    return minVal

def heuristic(emptyList):
    return sum(emptyList) #this heuristic function will return the sum of all stones in the piles

def ai_move_ida(emptyList):
    move = ida_star(emptyList)

    if move is not None:
        pile, stones = move
        print("\nNimX removes {} stones from pile {}.\n".format(stones, pile))
        emptyList[pile - 1] -= stones
        print("After NimX's turn,,,")
        continue_game(emptyList, "NimX")
    else:
        print("No valid move found for IDA*. This should not happen in a correct implementation.")
```

## ***Genetic Algorithm part***

```python
def init_population():
    population = [
        {"Algorithm": "Minimax", "Chromosome": minimaxChromo},
        {"Algorithm": "Alpha_Beta", "Chromosome": alphabetaChromo}
    ]
    return population

def calculate_fitness(chromosome, totalGames):
    wins = sum(chromosome) #total number of wins for each strategy will be used
    fitness = wins / totalGames if totalGames > 0 else 0
    return fitness

def fitness_function(population):
    populationFitness = []
    for item in population:
        algo = item["Algorithm"]
        chromo = item["Chromosome"]
        fitnessVal = calculate_fitness(chromo, len(chromo))
        print('\n')
        print(f"Algorithm: {algo}, Fitness: {fitnessVal}")
        populationFitness.append({
            "Algorithm": algo,
            "Chromosome": chromo,
            "Fitness_value": fitnessVal
        })
    return populationFitness

def selection(populationFitness):
    sortedFitness = sorted(populationFitness, key=lambda x: x["Fitness_value"], reverse=True)
    selected = sortedFitness[0]
    print("\n")
    print(f"Selected Chromosome: {selected['Chromosome']} with algotithm: {selected['Algorithm']}")
    return selected


def genetic_algorithm():
    population = init_population()
    populationFitness = fitness_function(population)
    selected = selection(populationFitness)
if __name__ == "__main__":
    main()
```
